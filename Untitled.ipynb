### Initial setup:

```python
## Import all modules:

import numpy as np
import random
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Set plot parameters:
plotclr = 'tab:orange'
plt.rcParams.update({'font.size' : 40})
plt.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['mathtext.fontset'] = 'cm'
clrs = ['#073b3a', '#0b6e4f', '#08a045', '#6bbf59', '#f8b464']
```

```python
## Parameters:

Delta = 0  # Detuning
Gamma = 1  # Decay rate
Omega = 10 * Gamma  # Rabi frequency
a, b = 1, 0  # Coefficients
psi = np.array([a, b])  # Two-level qubit

# Matrices:
I = np.array([[1, 0], [0, 1]])  # Identity matrix
H = 0.5 * np.array([[Delta, Omega], [Omega, -Delta]])  # Hamiltonian
H_eff = H - 0.5 * np.array([[0, 0], [0, 1j*Gamma]])  # Effective Hamiltonian

# Additional variables for simulation:
n = 1000  # Number of time steps
N = 500  # Number of runs
dt = 0.01  # Size of time step
t = np.zeros(n + 1)  # Array of timestamps
for i in range(1, n + 1):
    t[i] = i * dt
H_1 = I - 1j * H_eff * dt  # First-order Hamiltonian
rho = np.array([[abs(a)**2, a*np.conj(b)],
                [np.conj(a)*b, abs(b)**2]])  # Density matrix
```

```python
## Define methods.

# Renormalise qubit state:
def renorm(psi):
    mag = np.sqrt(abs(psi[0]**2) + abs(psi[1]**2))
    return psi / mag

# Update qubit state with new values of {a, b}:
def update(a, b):
    return np.array([a, b])

# Return magnitude of qubit state vector.
def magnitude(psi):
    return np.sqrt(abs(psi[0]**2) + abs(psi[1]**2))

# Set qubit to ground state.
def gst(psi):
    return np.array([1, 0])

# Set qubit to excited state.
def est(psi):
    return np.array([0, 1])

# Solve OBEs:
def solve(t):
    # Split into five terms (A-E) for simplicity.
    A = Omega ** 2 / (2 * Omega ** 2 + Gamma ** 2)
    B = np.exp(-(3 * Gamma * t) / 4)
    C = np.cos(np.sqrt(Omega ** 2 - (Gamma ** 2) / 16) * t)
    D = 3 * Gamma / np.sqrt(16 * Omega ** 2 - Gamma ** 2)
    E = np.sin(np.sqrt(Omega ** 2 - (Gamma ** 2) / 16) * t)
    return A * (1 - B * (C + D * E))
```

### Analytic Solution:

```python
## Solve OBE analytically.

# Create empty array to store values:
analytic = np.empty(n + 1)

# Solve equations and plot solution n times:
for i in range(n + 1):
    analytic[i] = solve(t[i])
    
# Plot analytic solution:
fig, ax = plt.subplots(figsize=(20, 12))
ax.plot(t, analytic, color=clrs[1], linewidth=3)
ax.plot([0, 10], [0.5, 0.5], linestyle='--', color=clrs[3], linewidth=3, alpha=0.5)
ax.set_xlabel('Time [1/{}]'.format(r'$\Gamma$'))
ax.set_ylabel('Excited state population')
ax.tick_params(direction="in", length=8)

plt.show()

fig.savefig('AnalyticOBE.pdf', bbox_inches='tight')
```

### Monte-Carlo Wavefunction:

```python
## Run MCW simulation N times.

arr = np.empty((N, n + 1))  # Quantum trajectory array

for i in range(N):
    psi = gst(psi)  # Initialise in ground state
    evol = [0]
    
    for j in range(n):
        psi = np.matmul(H_1, psi)  # Evolve state
        mag = magnitude(psi) ** 2  # Magnitude of state vector
        r = random.random()  # Generate random number between [0, 1]
        
        if r > mag:
            psi = gst(psi)  # Quantum jump occurred
        else:
            # Compute correction terms:
            corr_a = 0.5 * Gamma * dt * psi[1] ** 2
            corr_b = 0.5 * Gamma * dt * psi[0] ** 2
            psi[0] = psi[0] * (1 + corr_a)
            psi[1] = psi[1] * (1 - corr_b)
            psi = psi / np.sqrt(mag)
        evol += [abs(psi[1]) ** 2]
    arr[i] = evol
    
psi = gst(psi)
K = [0]

for k in range(n):
    psi = np.matmul(I - 1j * H * dt, psi)
    mag = magnitude(psi) ** 2
    psi = psi / np.sqrt(mag)
    K += [abs(psi[1]) ** 2]
```

```python
## Plot of a single quantum trajectory.

fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(t[:500], arr[8][:500], color=clrs[1], linewidth=3, label='$\Gamma = 1$')
ax.plot(t[:500], K[:500], color=clrs[3], alpha=0.5, linewidth=3, linestyle='--', label='$\Gamma = 0$')
ax.set_xlabel('Time [1/{}]'.format(r'$\Gamma$'))
ax.set_ylabel('Excited state\npopulation')
ax.tick_params(direction="in", length=8)
ax.legend(framealpha=1, loc=1)

plt.show()
fig.savefig('SingleTrajectoryMCW.pdf', bbox_inches='tight')
```

```python
## Comparison plot.

QT = arr.mean(axis=0)

# Calculate normalised residuals:
residuals = np.zeros(n + 1)
for i in range(n + 1):
    if analytic[i] == 0:
        residuals[i] = 0
    else:
        residuals[i] = (QT[i] - analytic[i]) / analytic[i]
        
# Generate figure:
fig, ax = plt.subplots(2, 1, figsize=(20, 16),
                       sharex=True, gridspec_kw={'height_ratios': [3, 1]})
plt.tight_layout()
ax[0].plot(t, QT, label='MCW', color=clrs[3], linewidth=3)
ax[0].plot(t, analytic, label='OBE', linestyle='--', linewidth=3, color=clrs[0])

# for j in range(int(N / 40)):
#     ax[0].plot(t, arr[j], color='0.8', alpha=0.1)

ax[1].plot(t, residuals, linewidth=3, color=clrs[3])
ax[1].plot([0, n * dt], [0, 0], linestyle='--', linewidth=3, color=clrs[0])

# Labels:
ax[0].set_ylabel('Excited state population')
ax[0].legend(framealpha=0.5, loc=1)
ax[1].set_xlabel('Time [1/{}]'.format(r'$\Gamma$'))
ax[1].set_ylabel('Normalised\nResiduals')

# Inward-facing ticks:
ax[0].tick_params(direction="in", length=8)
ax[1].tick_params(direction="in", length=8)

# Additional adjustments:
# plt.subplots_adjust(hspace=0)
# ax[1].set_ylim(min(residuals) * 1.5, max(residuals) * 1.5)

plt.show()

fig.savefig('Comparison.pdf', bbox_inches='tight')
```

### Vary Detuning:

```python
var_Delta = np.array([0, 5, 10, 20])
var_Delta_arr = np.zeros((N, n + 1))
VD = np.zeros((5, n + 1))

for h in range(4):
    Hml = 0.5 * np.array([[var_Delta[h], Omega],
                        [Omega, -var_Delta[h]]])
    Hml_eff = Hml - 0.5 * np.array([[0, 0], [0, 1j*Gamma]])
    Hml_1 = I - 1j * Hml_eff * dt
    
    for i in range(N):
        psi = gst(psi)
        evol = [0]
        
        for j in range(n):
            psi = np.matmul(Hml_1, psi)
            mag = magnitude(psi) ** 2
            r = random.random()
            if r > mag:
                psi = gst(psi)
            else:
                # Compute correction terms:
                corr_a = 0.5 * Gamma * dt * psi[1] ** 2
                corr_b = 0.5 * Gamma * dt * psi[0] ** 2
                psi[0] = psi[0] * (1 + corr_a)
                psi[1] = psi[1] * (1 - corr_b)
                psi = psi / np.sqrt(mag)
            evol += [abs(psi[1]) ** 2]
            
        var_Delta_arr[i] = evol
        
    VD[h] = var_Delta_arr.mean(axis=0)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.tight_layout()

for i in range(4):
    ax.plot(t, VD[i], label='$\Delta$ = {}'.format(var_Delta[i]), color=clrs[i], linewidth=3)   
ax.set_xlabel('Time [1/{}]'.format(r'$\Gamma$'))
ax.set_ylabel('Excited state population')
ax.tick_params(direction="in", length=8)
ax.legend(framealpha=0.5, ncol=4)    
plt.show()

fig.savefig('Detuning.pdf', bbox_inches='tight')
```

### Add Gaussian Noise to Parameters:

```python
## Generate normal distribution of noise.

noise = np.random.normal(0, 7, n)
Omegas = list(map(lambda x: x + Omega, noise))
Deltas = list(map(lambda x: x + Delta, noise))
```

```python
## Run evolution with noise once for Omega.

noise_Omega = np.empty(n + 1)
noise_Omega[0] = 0
psi = gst(psi)  # Initialise state

for i in range(n):
    Hml = 0.5 * np.array([[Delta, Omegas[i]], [Omegas[i], -Delta]])  # Apply noise to Hamiltonian
    Hml_eff = Hml - 0.5 * np.array([[0, 0], [0, 1j*Gamma]])
    Hml_1 = I - 1j * Hml_eff * dt
    psi = np.matmul(Hml_1, psi)
    mag = magnitude(psi) ** 2
    r = random.random()
    if r > mag:
        psi = gst(psi)
    else:
        corr_a = 0.5 * Gamma * dt * psi[1] ** 2
        corr_b = 0.5 * Gamma * dt * psi[0] ** 2
        psi[0] = psi[0] * (1 + corr_a)
        psi[1] = psi[1] * (1 - corr_b)
    
    psi = psi / np.sqrt(mag)  # Renormalise
    noise_Omega[i + 1] = abs(psi[1]) ** 2  # Append population of excited state to trajectory
```

```python
## Run evolution with noise once for Delta.

noise_Delta = np.empty(n + 1)
noise_Delta[0] = 0
psi = gst(psi)  # Initialise state

for i in range(n):
    Hml = 0.5 * np.array([[Deltas[i], Omega], [Omega, -Deltas[i]]])  # Apply noise to Hamiltonian
    Hml_eff = Hml - 0.5 * np.array([[0, 0], [0, 1j*Gamma]])
    Hml_1 = I - 1j * Hml_eff * dt
    psi = np.matmul(Hml_1, psi)
    mag = magnitude(psi) ** 2
    r = random.random()
    if r > mag:
        psi = gst(psi)
    else:
        corr_a = 0.5 * Gamma * dt * psi[1] ** 2
        corr_b = 0.5 * Gamma * dt * psi[0] ** 2
        psi[0] = psi[0] * (1 + corr_a)
        psi[1] = psi[1] * (1 - corr_b)
    
    psi = psi / np.sqrt(mag)  # Renormalise
    noise_Delta[i + 1] = abs(psi[1]) ** 2  # Append population of excited state to trajectory
```

```python
fig, ax = plt.subplots(2, figsize=(20, 12), sharex=True)

ax[0].plot(t[:500], noise_Omega[:500], label='$\Omega$', color=clrs[1], linewidth=3)
ax[0].plot(t[:500], K[:500], color=clrs[3], alpha=0.5, linewidth=3.5, linestyle='--')
ax[1].plot(t[:500], noise_Delta[:500], label='$\Delta$', color=clrs[1], linewidth=3)
ax[1].plot(t[:500], K[:500], color=clrs[3], alpha=0.5, linewidth=3.5, linestyle='--')

ax[0].set_ylabel('Excited state\npopulation')
ax[1].set_xlabel('Time [1/{}]'.format(r'$\Gamma$'))
ax[1].set_ylabel('Excited state\npopulation')

ax[0].tick_params(direction="in", length=8)
ax[1].tick_params(direction="in", length=8)

ax[0].legend(framealpha=1, loc=1)
ax[1].legend(framealpha=1, loc=1)
fig.tight_layout()

plt.show()

fig.savefig('GaussianNoise.pdf', bbox_inches='tight')
```
